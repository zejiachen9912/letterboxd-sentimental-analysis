## Aggregating 

```python
base_dir = "/Users/zejiachen/Desktop/fall 2021/QTM 340/Final Project/dataRating/"
# get a list of all the files in the directory
docs = os.listdir(base_dir)
```

```python
# getting the file name
filmName = [os.path.splitext(os.path.basename(x))[0] for x in docs]
filmName
```

```python
# read all the .csv file to dataframe
films = {}
for i in range(len(filmName)):
        films[filmName[i]] = pd.read_csv('/Users/zejiachen/Desktop/fall 2021/QTM 340/Final Project/dataRating/' + docs[i])
        
# adding the sentiment score and rating column
for i in filmName:
    films[i]['overall'] = 0.0
    films[i]['negative'] = ''
    films[i]['neutral'] = ''
    films[i]['positive'] = ''
    films[i]['compound'] = ''
    
```

```python
films['inception'][:8]
```

```python
cleanFilms = {}

for film in films:
# preprocess the data
    name = films[film]
    nrow = len(films[film].index)
    deleteRows = []
       
    for row in range(nrow):
        name['overall'][row] = f(name.iloc[row][1])
        if name.iloc[row][1].isascii() == True:
            deleteRows.append(name.iloc[row][0])
        elif len(name.iloc[row][2]) == 1:
            deleteRows.append(name.iloc[row][0])
    print(deleteRows)
    
    
    name = name.drop(deleteRows)
    name.reset_index(inplace=True) 
    name = name.drop(columns=['index', 'Unnamed: 0'])
    
    cleanFilms[film] = name
```

```python
# calculating the score for each of the comments
for film in filmName:
    name = cleanFilms[film]
    nrow = len(name.index)
    print('film: ' + film)
    print('nrow: ' + str(nrow))
    for i in range(nrow):
        scores = sid.polarity_scores(name['Comments'][i])
        name['negative'][i] = scores['neg']
        name['neutral'][i] = scores['neu']
        name['positive'][i] = scores['pos']
        name['compound'][i] = scores['compound']
    
```

### Aggregating  the dataset & Store it

```python
titleFilms = {}

for film in filmName:
    name = cleanFilms[film]
    copy = name.copy()
    copy['Film'] = film
    temp = pd.DataFrame(copy)
    titleFilms[film] = temp
```

```python
columns = ['Film', 'Rating', 'Comments', 'overall', 'negative', 'neutral', 'positive', 'compound']
filmReview = pd.DataFrame(columns = columns)
```

```python
for film in filmName:
    name = titleFilms[film]
    filmReview = pd.DataFrame(filmReview.append(name, ignore_index=True))
        
```
