## Aggregating

```python
base_dir = "/Users/zejiachen/Desktop/fall 2021/QTM 340/Final Project/finalProjectDataset/"
# get a list of all the files in the directory
docs = os.listdir(base_dir)
docs.pop(1)
```

```python
# getting the file name
filmName = [os.path.splitext(os.path.basename(x))[0] for x in docs]
```

```python
filmName
```

```python
# read all the .csv file to dataframe
films = {}
for i in range(len(filmName)):
        films[filmName[i]] = pd.read_csv('/Users/zejiachen/Desktop/fall 2021/QTM 340/Final Project/finalProjectDataset/' + docs[i])
        
# adding the sentiment score column
for i in filmName:
    films[i]['negative'] = ''
    films[i]['neutral'] = ''
    films[i]['positive'] = ''
    films[i]['compound'] = ''
    
```

```python
films['inception'][:10]
```

```python
# calculating the score for each of the comments
for film in filmName:
    nrow = len(films[film].index)
    print('film: ' + film)
    print('nrow: ' + str(nrow))
    for i in range(nrow):
        scores = sid.polarity_scores(films[film]['Comments'][i])
        films[film]['negative'][i] = scores['neg']
        films[film]['neutral'][i] = scores['neu']
        films[film]['positive'][i] = scores['pos']
        films[film]['compound'][i] = scores['compound']
    
```

## Tf-Idf

<p style="font-size:110%; font-family: courier">
    Tf-Idf model logic: <br> I am treating 
    each piece of review as a <b>document</b> whereas <br>
    the list of most posive/negative review for each Nolan's film as a <b>corpus </b> <br>
    <i>I don't know if I am doing this correctly</i>
<p>

```python
mostPosReviewsList = []
i = 0
for review in mostPosReviews:
    mostPosReviewsList.append(mostPosReviews[review]['Comments'].tolist())
    i += 1
    
mostNegReviewsList = []
j = 0
for review in mostNegReviews:
    mostNegReviewsList.append(mostNegReviews[review]['Comments'].tolist())
    j += 1
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

### Positive Matrix

```python
# create matrix for each of the film
tfidf_vectorizer=TfidfVectorizer(stop_words='english', use_idf=True) # excludings stopwords again
 

mostPosReviewsMatrix = {}
x = 0
for i in mostPosReviewsList:
    tfidf_vectorizer_vectors=tfidf_vectorizer.fit_transform(i)
    mostPosReviewsMatrix[filmName[x]] = pd.DataFrame(tfidf_vectorizer_vectors.toarray(), columns=tfidf_vectorizer.get_feature_names())
    x += 1
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)


for film in filmName:
    print(film)
    display(mostPosReviewsMatrix[film][:5])
```

```python
top_dict_pos = {}
for film in filmName:
    data = mostPosReviewsMatrix[film].transpose()
    for c in range(49):
        top = data.iloc[:,c].sort_values(ascending = False).head(30)
        top_dict_pos[film] = list(zip(top.index, top.values))
```

```python
# Print the top 15 words said by positive reviews
for film, top_words in top_dict_pos.items():
    print(film)
    print(', '.join([word for word, count in top_words[0:9]]))
    print('---')
```

### Negative Matrix

```python
mostNegReviewsMatrix = {}
y = 0
for i in mostNegReviewsList:
    tfidf_vectorizer_vectors=tfidf_vectorizer.fit_transform(i)
    mostNegReviewsMatrix[filmName[y]] = pd.DataFrame(tfidf_vectorizer_vectors.toarray(), columns=tfidf_vectorizer.get_feature_names())
    y += 1
    
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)


for film in filmName:
    print(film)
    display(mostNegReviewsMatrix[film][:5])
```

```python
top_dict_neg = {}
for film in filmName:
    data = mostNegReviewsMatrix[film].transpose()
    for c in range(49):
        top = data.iloc[:,c].sort_values(ascending = False).head(30)
        top_dict_neg[film] = list(zip(top.index, top.values))
```

```python
# Print the top 10 words said by positive reviews
for film, top_words in top_dict_neg.items():
    print(film)
    print(', '.join([word for word, count in top_words[0:9]]))
    print('---')
```

## Insights & Findings

<p style="font-size:120%;">For this model testing assignment, I conducted a sentimental analysis in respect to each Nolan's film reviews using VADER. <br>
Depending on the sentimental score, I then create a series of most positive/negative reviews for each of Nolan's film, and eventually leverage Tf-Idf to analyze the significance of terms among a film review document. <p>

<p style="font-size:120%;">Looking at some of the strongest positive/negative reviews in <code>mostPosReviews</code> and <code>mostNegReviews</code>, VADER does a great job in identifying sentiments from all reviewers, espeicially positive sentiments. Words like "masterpiece" and "brilliant" all are great indicators of a person's love toward the film. <br>
To my surprise, VADER also does a great job in indentifying positive sentiments toward actors or composers. One of the postivie review in inception given by VADER is "jgl sexy." By "jgl," I am assuming the reviewer is alluding the actor Joseph Gordon-Levitt playing in <i>Inception</i>. <p>

<p style="font-size:120%;">Nevertheless, I realized that VADER does not do a great job in identifying negative sentiments. For almost every film, there is positive review being treated as negative one by VADER. <br>
For example, the most negative review in <i> Inception </i> identified by VADER is "FUCK IT FIVE STARS." We also have review like "Fucking INSANE." being wrongly treated as negative by VADER. <br>
I suspect that this might attribute to the sustanial amount of swear words that reviewers are using. (Should I clean the swear word or there are other approahces that I can resolve this bias?) <br>
What is more, VADER does not seem to correctly detect a sense of loss or condolence expressed by reviewers. Oftentime, when expressing a sense of loss to the actor who passed away in a film, instead of critizing it, the review are actually demostrating their appreciation to the film, memorizing the actor's repertoire. VADER, on the other hand, cannot detecting this layer of sentiment from the reviwer. As a result, we see reviews like "I miss poor Robin Williams" and "UGH I MISS HEATH SO MUCH." <p>

<p style="font-size:120%;"> After using VADER for calculating sentimental score, I decided to yet conduct another analysis utilizing Tf-Idf, with a aim of defining important terms inside a film review document. <br>
I printed out 15 most significant words said by positive/negative reviews in a particular film. I will say the result is solid. There are quite some words can define the seemingly define the characristic of a film. <br>
For example, "protector", "soldier", "defender" in <i> Dunkirk </i>, "superhero" in <i> Batman, The Dark Knight </i> 
We can see the some the word in the corpus does have the ability to entail characteristic of a certain film which is quite interesting. <br>
