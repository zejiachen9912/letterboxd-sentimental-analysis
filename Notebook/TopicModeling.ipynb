## Topic Modeling

```python
stop_words = stopwords.words('english')
stop_words.extend(['inception', 'inception', 'tenet', 'memento', 'prestige', 'following', 'insomnia', 'dunkirk', 'nolan', 'christopher'])
```

```python
# tokenizing the text and removing stopwords
def tokenize(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
        
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]


data = allFilmsLower.Comments.values.tolist()
stream = tokenize(data)
# remove stop words
stream = remove_stopwords(stream)
```

```python
import gensim.corpora as corpora
# Create Dictionary
id2wordReviews = corpora.Dictionary(stream)
print(id2wordReviews)
```

```python
count = 0
for i in id2wordReviews.token2id:
    print(i + ": " + str(id2wordReviews.token2id[i]))
    count += 1
    if count == 10:
        break
```

```python
# this line filters out words that appear only 1 doc, keeping the rest
# I will use this one 
# note how no_below and no_above take different values
id2wordReviews.filter_extremes(no_below=2, no_above=1.0)

print(id2wordReviews)
```

```python
# Create Corpus
texts = stream
# Term Document Frequency
corpus = [id2wordReviews.doc2bow(text) for text in texts]
```

```python
corpus[:5]
```

```python
from pprint import pprint
# number of topics
num_topics = 9
# Build LDA model
film_lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2wordReviews,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
pprint(film_lda_model.print_topics())
doc_lda = film_lda_model[corpus]
```

```python
# let's format the words a little more nicely; 
# the formatted=False parameter returns tuples of (word, probability)

topics = film_lda_model.show_topics(15, 20, formatted=False)

for topic in topics:
    topic_num = topic[0]
    topic_words = ""
    
    topic_pairs = topic[1]
    for pair in topic_pairs:
        topic_words += pair[0] + ", "
    
    print("T" + str(topic_num) + ": " + topic_words)
```
